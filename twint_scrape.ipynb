## Get all of my Hashtags

```python
hashtags = []

def hash_getter(tweet):
    global hashtags
    
    for tag in tweet:
        if tag not in hashtags:
            hashtags.append(tag)        
    
```

```python
Tweets_df['hashtags'].apply(hash_getter)
```

```python
import pickle
# for writing
with open('hashtags', 'wb') as fp:
    pickle.dump(hashtags, fp)
    
# for reading
with open ('hashtags', 'rb') as fp:
    list_1 = pickle.load(fp)
```

## Clean Tweets

```python
import re
```

```python
# Take out HTML
pattern = r"http\S+|pic\.\S+|\xa0|‚Ä¶"
# Take out Hashtags
pattern += r"|#[a-zA-Z0-9_]+"
```

```python
def tweet_stripper(tweet):
    # completely erase the HTML and hashtags
    tweet = re.sub(pattern, "", tweet).strip()
    # just get rid of the @ symbols before a person's name
    tweet = tweet.replace('@','')
    return tweet
```

```python
Tweets_df['fixed_tweet'] = Tweets_df['tweet'].apply(tweet_stripper)
```
